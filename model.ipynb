```python
import numpy as np
import pandas as pd
from sklearn .model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python

import seaborn as sns
df= sns.load_dataset("titanic")
df

```

```python

df.drop(columns=['class','who','adult_male','deck','embark_town','alive','alone'],inplace=True)
```

```python

# srep 1 train/test/split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['survived']),df['survived'],test_size=0.2,random_state=42)

X_train.head()

y_train.sample(5)


Create Pipeline
```

```python

# imputation transformer
trf1=ColumnTransformer([('impute_age',SimpleImputer(),[2]),('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])],remainder='passthrough')

#one hot encoding
trf2=ColumnTransformer([('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])],remainder='passthrough')

#scaling
trf3=ColumnTransformer([('scale',MinMaxScaler(),slice(0,10))])

# feature selection
trf4=SelectKBest(score_func=chi2,k=8)

# train the model
trf5=DecisionTreeClassifier()
```

```python

pipe=Pipeline(
    [("trf1",trf1),
    ("trf2",trf2),
    ("trf3",trf3),
    ("trf4",trf4),
    ("trf5",trf5)]

)

#train
pipe.fit(X_train,y_train)



import pickle
import numpy as np

f= open("pipe.pkl","wb")
pickle.dump(pipe,f)
```
